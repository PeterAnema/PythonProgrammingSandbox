```python
import numpy as np
```

```python
np.arange(10)
```

```python
x = np.arange(0,10,0.1)
x
```

```python
y = np.sin(x)
y
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot(x,y)
plt.show()
```

```python
fig, ax = plt.subplots(2,2)
ax[0,0].plot(x,y)
ax[0,1].plot(x,y+3)
ax[1,1].plot(x,x)
ax[1,0].plot(x,np.tan(x))

plt.show()
```
